```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
iris
```

```python
#CONVERTING DATA TO DATAFRAME 
```

```python
iris_df = pd.DataFrame(np.c_[iris['data'],iris['target']], 
                       columns = np.append(iris['feature_names'],['target']))
```

```python
iris_df
```

```python
pd.set_option('display.max_rows',None)
pd.set_option('display.max_columns',None)
```

```python
iris_df
```

```python
sn.pairplot(iris_df)
```

```python
sn.pairplot(iris_df, vars = ['sepal length (cm)',
  'sepal width (cm)',
  'petal length (cm)',
  'petal width (cm)'])
```

```python
sn.pairplot(iris_df, vars = ['sepal length (cm)',
  'sepal width (cm)',
  'petal length (cm)',
  'petal width (cm)'], hue = 'target')
```

```python
sn.pairplot(iris_df, vars = ['sepal length (cm)',
  'sepal width (cm)',
  'petal length (cm)',
  'petal width (cm)'], hue = 'target', hue_order = [1.0,0.0])
```

```python
sn.pairplot(iris_df, vars = ['sepal length (cm)',
  'sepal width (cm)',
  'petal length (cm)',
  'petal width (cm)'], hue = 'target', palette = 'Dark2')
```

```python
sn.pairplot(iris_df, hue = 'target', x_vars = ['sepal length (cm)','sepal width (cm)'], y_vars = ['petal length (cm)',
  'petal width (cm)'])
```

```python
sn.pairplot(iris_df, vars = ['sepal length (cm)',
  'sepal width (cm)',
  'petal length (cm)',
  'petal width (cm)'], hue = 'target', kind = 'reg')
```

```python
sn.pairplot(iris_df, vars = ['sepal length (cm)',
  'sepal width (cm)',
  'petal length (cm)',
  'petal width (cm)'], hue = 'target', diag_kind = 'kde')
```

```python
sn.pairplot(iris_df, vars = ['sepal length (cm)',
  'sepal width (cm)',
  'petal length (cm)',
  'petal width (cm)'], hue = 'target', markers = '*')
```

```python
sn.pairplot(iris_df, vars = ['sepal length (cm)',
  'sepal width (cm)',
  'petal length (cm)',
  'petal width (cm)'], hue = 'target', height = 2.5)
```

```python
sn.pairplot(iris_df, vars = ['sepal length (cm)',
  'sepal width (cm)',
  'petal length (cm)',
  'petal width (cm)'], hue = 'target', aspect =5)
```
